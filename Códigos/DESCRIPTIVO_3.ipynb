```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Cargar el archivo Excel
df = pd.read_excel('df_diario_promediado.xlsx')
```

```python

# Convertir la columna 'Dia' a datetime si no lo está ya
df['Dia'] = pd.to_datetime(df['Dia'])

# Configurar el estilo del gráfico
plt.figure(figsize=(15, 6))

# Crear el gráfico de línea
plt.plot(df['Dia'], df['TemperaturaPromedio'],
         linewidth=0.8, alpha=0.7, color='blue')

# Personalizar el gráfico
plt.title('Serie Temporal de Temperatura Promedio Diaria (2006-2024)',
          fontsize=14, fontweight='bold')
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Temperatura Promedio (°C)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')

# Formatear el eje X para mostrar años
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

# Rotar las fechas para mejor visualización
plt.xticks(rotation=45)

# Ajustar layout
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
